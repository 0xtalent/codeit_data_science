```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\hayong\Downloads\laptops.csv')
df
```

```python
brand_nation = {
    'Dell': 'U.S.',
    'Apple': 'U.S.',
    'Acer': 'Taiwan',
    'HP': 'U.S.',
    'Lenovo': 'China',
    'Alienware': 'U.S.',
    'Microsoft': 'U.S.',
    'Asus': 'Taiwan'
}
```

```python
df['brand'].map(brand_nation)
```

```python
df['brand_nations'] = df['brand'].map(brand_nation)
df
```

```python
# 200226 13:47
# 새로운 인사이트 발견하기
# 08. groupby
```

```python
nation_groups = df.groupby('brand_nations')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```

```python
# groupby(col)은 col의 값이 같은 row들을 group으로 묶는다고 생각하시면 됩니다. groupby 오브젝트에 .mean(), sum()와 같은 함수를 적용하면 모든 컬럼에 대한 group의 평균 / 합을 구하게 됩니다. 예를 들어 df.groupby('occupation').sum() 은 직업별 그 직업을 가지고 있는 모든 사람의 나이의 합을 구합니다 (gender 컬럼은 합을 구할 수 없습니다).
```
