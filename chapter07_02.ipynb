```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\hayong\Downloads\young_survey.csv')
df.head()
```

```python
interests = df.loc[:, 'History':'Pets']
# 모든 로우에 대해서 history 컬럼부터 pets 컬럼까지
interests.head()
```

```python
corr = interests.corr()
corr
```

```python
corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(corr)
```
